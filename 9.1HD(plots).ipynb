```python
import matplotlib.pyplot as plt
```

```python
n_anon = ['3', '7', '8', '9', '10', '17', '19', '21', '27', '31']
scores = [0, 0.00015, 0.00085, 0.00206, 0.00657, 0.03020, 0.03182, 0.04909, 0.05043, 0.05621]
sq_err = [100, 99.99045, 99.95162, 99.85344, 99.54842, 99.02178, 98.94182, 98.26478, 99.06461, 98.59320]

fig, ax = plt.subplots(figsize= [8, 7])
ax.plot(n_anon, scores, color='blue', marker='o', linewidth=2)
ax.set_xlabel('k Values', fontsize=15)
ax.set_ylabel('Utility Loss',  color='blue', fontsize=15)
ax2 = ax.twinx()
ax2.plot(n_anon, sq_err, color='red', marker='o', linewidth=2)
ax2.set_ylabel('Mean Squared Error (%)',  color='red', fontsize=15)
plt.title("k-Anonymity with various parameters", fontsize=20)
plt.show()
```
